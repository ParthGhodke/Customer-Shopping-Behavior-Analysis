# Problem Statement
A leading retail company wants to better understand its customers shopping behaviour in order to improve sales, customer satisfaction, and long term loyalty. The management team has noticed changes in purchasing patterns across demogaphics, product categories and sales chanels(online and offline). They are particularly intersted in uncovering which factors suchs as discounts, reviews, seasons, or payment preferences, drive consumer decisions and repeat purchases.

you are tasked with analyzing the company's consumer behaviours dataset to answer the following overarching business question:
**"How can the Company leverage consumer shopping data to identify trend,improve customer engagement, and optimize marketing and product strategies?"**

```python
import pandas as pd
df = pd.read_csv('/content/customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include= 'all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#Creating another feature called column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

Connecting Python script to database

```python
import sqlalchemy
import pandas as pd

# Ensure the DataFrame 'df' is loaded from previous steps
# If this cell is run independently, and df is not defined, it would cause an error.
# Assuming 'df' is already in the global scope from previous successful runs.

# Create a SQLite database engine.
# This will create or connect to a file named 'customer_data.db' in your current working directory.
engine = sqlalchemy.create_engine('sqlite:///customer_data.db')
print("SQLite database engine created successfully.")
```

Now, we will write your `df` DataFrame into a table named `customer_shopping_data` within the `customer_data.db` SQLite database.

```python
# Define the table name for your DataFrame in SQLite
table_name = "customer_shopping_data"

# Write the DataFrame to the SQLite table.
# 'if_exists='replace'' will overwrite the table if it already exists.
# 'index=False' prevents pandas from writing the DataFrame index as a column in the database table.
df.to_sql(table_name, con=engine, if_exists='replace', index=False)

print(f"DataFrame successfully written to table '{table_name}' in customer_data.db")
```

To confirm that the data was written correctly, we can read it back from the SQLite database into a new pandas DataFrame.

```python
# Read the data back from the SQLite database into a new DataFrame
df_from_db = pd.read_sql(table_name, con=engine)

display(df_from_db.head())
print("Data successfully read back from SQLite database.")
```
